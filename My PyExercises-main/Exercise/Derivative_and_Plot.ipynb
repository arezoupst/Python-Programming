```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')

# Define the function
f = sp.log(x)

# Calculate the derivative
df_dx = sp.diff(f, x)

# Convert the functions to lambda functions for easy plotting
f_lambda = sp.lambdify(x, f, 'numpy')
df_dx_lambda = sp.lambdify(x, df_dx, 'numpy')

# Create a range of x values
x_vals = np.linspace(0.1, 10, 100)

# Evaluate the functions on the x values
F = f_lambda(x_vals)
DF_DX = df_dx_lambda(x_vals)

# Plot the functions
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(x_vals, F)
axs[0].set_title('Function')
axs[1].plot(x_vals, DF_DX)
axs[1].set_title('Derivative')
plt.show()
```
